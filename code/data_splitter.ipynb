# Data Splitter

```python
import pandas as pd
import re

init_data = pd.read_csv('../data/new_dataset.csv', encoding='utf-8')
print(init_data.columns)

#Title does not need cutoff, as the titles have a significantly shorter length
init_title_class_data = init_data.copy()

avg_text_length = init_data['text'].dropna().apply(len).mean()
avg_title_length = init_data['title'].dropna().apply(len).mean()
print(avg_text_length)
print(avg_title_length)

cutoff_value = int(avg_text_length) + int(avg_title_length)
print(cutoff_value)

```

```python
# Define function to clean whitespace
def clean_text(text):
    text = re.sub(r'<.*?>', '', text)  # remove HTML
    text = re.sub(r'http\S+', '', text)  # remove URLs
    text = re.sub(r'[^a-zA-Z0-9\s]', '', text)  # remove non-letters & keep numbers 
    return text.strip()

#creating column with merged title and text
class_data = init_data.copy()
class_data['input'] = class_data['title'] + ' ' + class_data['text']
class_data.drop('title', axis=1, inplace=True)
class_data.drop('text', axis=1, inplace=True)
class_data.drop('Unnamed: 0', axis=1, inplace=True)

#remove empty cells
class_data = class_data.replace('', pd.NA).dropna()
class_data['input'] = class_data['input'].apply(clean_text)

#remove cells, which have more than the cutoff value
class_data = class_data[class_data['input'].dropna().apply(len) <= cutoff_value]
class_data.dropna(subset=['input'])
class_data['input'] = class_data['input'].fillna('')


#print(class_data.columns)
#print(class_data)
#print(class_data[class_data['label'] == 1])
avg_input_length = class_data['input'].dropna().apply(len).mean()
#print(avg_input_length)

class_data.to_csv('../data/new_class_dataset.csv', index=False)

title_class_data = init_title_class_data.copy()
title_class_data.drop('text', axis=1, inplace=True)
title_class_data.drop('Unnamed: 0', axis=1, inplace=True)
title_class_data = title_class_data.replace('', pd.NA).dropna()
title_class_data['title'] = title_class_data['title'].apply(clean_text)
print(title_class_data.columns)

title_class_data.to_csv("../data/title_class_dataset.csv", index=False)


print("successfully saved data to csv!")
```
